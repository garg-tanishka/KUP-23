```python
import pytesseract
import PIL.Image
```

```python
unstructured = pytesseract.image_to_string(PIL.Image.open("unstructured.png"))
print(unstructured)


file_path = 'output.txt'  
with open(file_path, 'w') as file:
    file.write(unstructured)


print(f"Extracted text has been saved to '{file_path}'")
```

```python
import json


data = {
    "text": unstructured
}

json_data = json.dumps(data)


file_path = 'output.json'  
with open(file_path, 'w') as file:
    file.write(json_data)


print(f"Extracted text has been saved to '{file_path}' in JSON format.")
```

```python

from PIL import Image
import pyocr
import pyocr.builders 
```

```python
tools = pyocr.get_available_tools()
print(len(tools))

 

tool = tools[0]
print("Tools: ", tool.get_name())
print("Available languages: ", tool.get_available_languages()) 
```

```python
#unstructured image
path = 'unstructured.png'
img = Image.open(path)

txt = tool.image_to_string(
    img,
    lang='eng',
    builder=pyocr.builders.TextBuilder()
)
 
import re
normalized_text = re.split(r'\s+|[,!]', txt)  # Replace multiple line breaks with a single space
# normalized_text = re.sub(r'\s+', ' ', normalized_text)
print(normalized_text)
```

```python
#structured images
path = 'structured2.jpeg'
img = Image.open(path)

txt = tool.image_to_string(
    img,
    lang='eng',
    builder=pyocr.builders.TextBuilder()
)
 

print(txt)
```
