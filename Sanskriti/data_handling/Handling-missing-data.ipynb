```python
import pandas as pd
import numpy as np

nfl_data = pd.read_csv("NFL.csv")
nfl_data

#seed function for taking random values 
np.random.seed(0)
```

```python
nfl_data.head()
```

```python
#get the number of missing data points
missing_values_count = nfl_data.isnull().sum()
```

```python
missing_values_count
```

```python
missing_values_count[0:10]  #seeing the first 10 values
```

```python
#total missing values 
total_cells = np.product(nfl_data.shape)
total_cells
```

```python
total_missing = missing_values_count.sum()
total_missing
```

```python
#percentage of data missing 
percent_missing = (total_missing/total_cells)*100
percent_missing
```

```python
#approach 1 
# drop missing values
nfl_data.dropna()  #removed all the data becoz every coulmn atleast contain one nan value
```

```python
# remove all column with atleast one missing value
columns_with_na_dropped = nfl_data.dropna(axis=1)
columns_with_na_dropped.head()
```

```python
#how much data we actually lost
print("Coulmn in original data set : %d \n" %nfl_data.shape[1])
print("Coulmn with na's dropped : %d" %columns_with_na_dropped.shape[1])
```

```python
#approach 2
# filling in missing values automatically
#get a small subset of the NFL dataset
subset_nfl_dataset = nfl_data.loc[:,'penalty_player_id':'defensive_extra_point_conv'].head()
subset_nfl_dataset
```

```python
# filling nan values with 0
subset_nfl_dataset.fillna(0)
```

```python
 #replace all NA's the value that comes directly after it in the same column, 
# then replace all the remaining na's with 0
subset_nfl_dataset.fillna(method='bfill',axis =0).fillna(0)
```
